```python
coustomer_reviews =  ['sam was a great help to me in the store', 
                    'the cashier was very rude to me, I think her name was eleanor', 
                    'amazing work from sadeen!', 
                    'sarah was able to help me find the items i needed quickly', 
                    'lucy is such a great addition to the team', 
                    'great service from sara she found me what i wanted'
                   ]
```

```python
person_name = r"sarah?"
```

```python
reviews_results = []
```

```python
for review in coustomer_reviews:
    if(re.search(person_name, review)):
        reviews_results.append(review)
        
```

```python
print(reviews_results)
```

```python
a_reviews = []
a_pattern_to_find = r"^a"
for itms in coustomer_reviews:
    if(re.search(a_pattern_to_find, itms)):
        a_reviews.append(itms)
```

```python
print(a_reviews)
```

```python
y_reviews = []
y_pattern_to_find = r"y$"
for itms in coustomer_reviews:
    if(re.search(y_pattern_to_find, itms)):
        y_reviews.append(itms)
```

```python
print(y_reviews)
```

```python
needwant_reviews = []
pattern_to_find = r"(need|want)ed"
for itms in coustomer_reviews:
    if(re.search(pattern_to_find, itms)):
        needwant_reviews.append(itms)
print(needwant_reviews)
```
